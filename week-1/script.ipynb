### Cleaning using fillna / dropna

```python
df.isnull().sum()
```

```python
df["discount"]
```

```python
df["discount"] = df["discount"].fillna(0.0)
```

```python
df['what_customers_said']
```

```python
df['what_customers_said'] = df['what_customers_said'].fillna('')
```

```python
df['seller_name']
```

```python
df['seller_name'] = df['seller_name'].fillna('')
```

```python
df['videos'].head()
```

```python
df['videos'] = df['videos'].apply(lambda x: x if isinstance(x, list) else [])
```

```python
df['seller_information'].unique()
```

```python
df['seller_information'] = df['seller_information'].fillna('')
```

```python
df['variations']
```

```python
df['variations'] = df['variations'].apply(lambda x: x if isinstance(x, list) else [])
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

### Transformations

```python
df[['product_id', 'title', 'product_description', 'rating']]
```

```python
df[['product_id', 'title', 'product_description', 'rating']][df.rating > 4.0].sort_values(by='rating')
```

```python
df[['product_id', 'title', 'rating']][df.rating > 4.5].sort_values(by='product_id').head()
```

```python
df = df.sort_values(by=['product_id', 'rating'])
```

### Dropping Duplicates

```python
df.drop_duplicates(subset=['product_id'], inplace=True)
```
